```python
#OpenCV is for workign with images 
# Enter into terminal (pip2 install opencv)
# Check in terminal window and enter pip3 list 

import cv2
import numpy as np 
import matplotlib.pyplot as plt 
cv2.__version__
```

```python
#Loading Images 

img1= cv2.imread('/Users/arianauwaibi/Library/Mobile Documents/com~apple~CloudDocs/Graduate School Files/Semesters/Spring 2022/Machine Learning Course /Homework:Documents/images_for_python/forest.jpg')
img2= cv2.imread('/Users/arianauwaibi/Library/Mobile Documents/com~apple~CloudDocs/Graduate School Files/Semesters/Spring 2022/Machine Learning Course /Homework:Documents/images_for_python/people-walking.jpg')
type(img1)
print('showing the size of image 1',img1.shape)
print('showing the size of image 2',img2.shape)
img1_reshape = cv2.resize(img1,(300,300), interpolation = cv2.INTER_AREA)
img2_reshape = cv2.resize(img2,(300,300), interpolation = cv2.INTER_AREA)
print('Showing size of img2_reshaped',img2_reshape.shape)
```

```python
#Printing the Images 
plt.figure(figsize=(10,10))

plt.subplot(121)
plt.imshow(img1_reshape)

plt.subplot(122)
plt.imshow(img2_reshape)
```

```python
#Printing RGB channels 
plt.figure(figsize=(10,10))

plt.subplot(131)
plt.imshow(img1_reshape[:,:,0])
plt.gca().set_title('Red Channel')

plt.subplot(132)
plt.imshow(img1_reshape[:,:,1])
plt.gca().set_title('Green Channel')

plt.subplot(133)
plt.imshow(img1_reshape[:,:,2])
plt.gca().set_title('Blue Channel')
```

```python
#Define New Varaible the images are being added 

dst= cv2.addWeighted(img2_reshape,0.1,img1_reshape,0.5,0.1)

# the weifght of the two images added are controlled by the parameters (0.1...)

plt.figure(figsize = (10,10))
plt.imshow(dst)
```

```python
#Subtract two images 
dst_sub = cv2.subtract(img1_reshape,img2_reshape)
plt.figure(figsize = (10,10))
plt.imshow(dst_sub)
```

```python
#Define 

img_cat = cv2.imread('/Users/arianauwaibi/Library/Mobile Documents/com~apple~CloudDocs/Graduate School Files/Semesters/Spring 2022/Machine Learning Course /Homework:Documents/images_for_python/cat.jpg')
negative = 255 - img_cat
#goes for each element of the matrix and subtract 255
#the range of the image is 0-255 (.jpg color image)
# grayscale image 0-1

plt.figure(figsize = (10,10))

plt.subplot(121)
plt.imshow(img_cat)
plt.gca().set_title('Orginal Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(negative)
plt.gca().set_title('Negative Image')
plt.axis('off')
```

```python
img_cat.shape
```

```python
#Blurring the Image 
#Called image convolution
#the 40,40 is a filter / bin to filter the image in small squares rather than the entire image at once. 

img_blur = cv2.blur(img_cat,(40,40))
img_filter = cv2.boxFilter(img_cat,-5,(30,30),normalize= True)
img_gaussianfilter = cv2.GaussianBlur(img_cat,(41,41), cv2.BORDER_DEFAULT)


plt.figure(figsize= (20,20))

plt.subplot(141)
plt.imshow(img_cat)
plt.gca().set_title('Original Image')
plt.axis('off')

plt.subplot(142)
plt.imshow(img_blur)
plt.gca().set_title('Blur Image')
plt.axis('off')

plt.subplot(143)
plt.imshow(img_filter)
plt.gca().set_title('Filter Image')
plt.axis('off')

plt.subplot(144)
plt.imshow(img_gaussianfilter)
plt.gca().set_title('Guassian Image')
plt.axis('off')
```

```python
#differnt method to blur image 

img_median = cv2.medianBlur(img_cat,31)

plt.figure(figsize=(10,10))

plt.subplot(121)
plt.imshow(img_cat)
plt.gca().set_title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(img_median)
plt.gca().set_title('Mediam Blur')
plt.axis('off')
```

```python
#Transform RGB image into Greyscale or 

img_cat_hsv = cv2.cvtColor(img_cat,cv2.cv2.COLOR_RGB2HSV)
plt.imshow(img_cat_hsv)
```

```python
#Define Segementation

upper_limit = (119,119,13)
lower_limit = (191,191,61)

mask = cv2.inRange(img_cat,upper_limit,lower_limit)

result = cv2.bitwise_and(img_cat,img_cat, mask=mask)

plt.figure(figsize=(20,20))

plt.subplot(121)
plt.imshow(mask, cmap= 'gray')

plt.subplot(122)
plt.imshow(result)
```
