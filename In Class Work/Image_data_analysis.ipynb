```python
#OpenCV is for workign with images 
# Enter into terminal (pip2 install opencv)
# Check in terminal window and enter pip3 list 

import cv2
import numpy as np 
import matplotlib.pyplot as plt 
cv2.__version__
```

```python
#Loading Images 

img1= cv2.imread('/Users/arianauwaibi/Library/Mobile Documents/com~apple~CloudDocs/Graduate School Files/Semesters/Spring 2022/Machine Learning Course /Homework:Documents/images_for_python/forest.jpg')
img2= cv2.imread('/Users/arianauwaibi/Library/Mobile Documents/com~apple~CloudDocs/Graduate School Files/Semesters/Spring 2022/Machine Learning Course /Homework:Documents/images_for_python/people-walking.jpg')
type(img1)
print('showing the size of image 1',img1.shape)
print('showing the size of image 2',img2.shape)
img1_reshape = cv2.resize(img1,(300,300), interpolation = cv2.INTER_AREA)
img2_reshape = cv2.resize(img2,(300,300), interpolation = cv2.INTER_AREA)
print('Showing size of img2_reshaped',img2_reshape.shape)
```

```python
#Printing the Images 
plt.figure(figsize=(10,10))

plt.subplot(121)
plt.imshow(img1_reshape)

plt.subplot(122)
plt.imshow(img2_reshape)
```

```python
#Printing RGB channels 
plt.figure(figsize=(10,10))

plt.subplot(131)
plt.imshow(img1_reshape[:,:,0])
plt.gca().set_title('Red Channel')

plt.subplot(132)
plt.imshow(img1_reshape[:,:,1])
plt.gca().set_title('Green Channel')

plt.subplot(133)
plt.imshow(img1_reshape[:,:,2])
plt.gca().set_title('Blue Channel')
```
