```python
# Create a program that when it is exceuted show is the number of even (2,4,6)
#numbers obtained in 10,000 dice roll

# Define an array with all possible outcomes 
possible_outcomes = [1,2,3,4,5,6]

#Define the counter of the times an event number was obtained 
number_of_executation = 10000

# wanted outcome 
wanted_outcomes = [2,4,6]

```

```python
import numpy as np

number_of_even= 0 

# We will write an iteration for the dice roll 

for i in range(number_of_executation): 
    x= np.random.choice(possible_outcomes)
    if (x in wanted_outcomes): 
        number_of_even += 1 
```

```python
# Compute the Probablity 

probability_of_even = number_of_even/ number_of_executation
print('{0:d} of the outcomes were even, we had a proability of around {1:.2f}'.format(number_of_even,probability_of_even))
#0 the 0 postion 
# d is double (an interger)

#1 means 
#2f means 
```
